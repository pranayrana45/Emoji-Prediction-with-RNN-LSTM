## Processing Data

```python
import pandas as pd
train = pd.read_csv('train_emoji.csv',header = None).values
test = pd.read_csv('test_emoji.csv',header= None).values
```

```python
x_train = train[:,0]
y_train = train[:,1]

x_test = test[:,0]
y_test = test[:,1]
print(y_test)
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train,num_classes = 5)
```

```python
y_test = to_categorical(y_test,num_classes = 5)
```

```python
# Reading Glove vetors file
f = open(r"C:\Users\prana\OneDrive\Desktop\Emoji Prediction\glove\glove.6B.50d.txt",encoding='utf-8')
```

```python
import numpy as np
embedding = {}
for line in f:
    value = line.split()
    word = value[0]
    coef = np.asarray(value[1:],dtype = 'float')
    embedding[word] = coef
f.close()
```

```python
# Embedding using 
def embedding_output(X):
    embedding_out = np.zeros((X.shape[0],10,50))
    
    for i in range(X.shape[0]):
        X[i] = X[i].split()
        for j in range(len(X[i])):
            try:
                embedding_out[i][j] = embedding[X[i][j].lower()]
            except:
                embedding_out[i][j] = np.zeros((50,))
                
    return embedding_out
```

```python
embedding_output_train= embedding_output(x_train)
embedding_output_test = embedding_output(x_test)
```

## Model Training

```python
from keras.models import Sequential
from keras.layers import *
```

```python
model = Sequential()
model.add(LSTM(64,input_shape = (10,50)))
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',optimizer = 'adam', metrics=['accuracy'])
model.summary()
```

```python
hist = model.fit(embedding_output_train,y_train,epochs = 50, batch_size = 64, shuffle = True, validation_split=0.2)
```

## Accuracy

```python
model.evaluate(embedding_output_test,y_test)
```

## Making Predictions

```python
pred = model.predict_classes(embedding_output_test)
```
